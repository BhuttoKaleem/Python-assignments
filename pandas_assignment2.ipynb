```python
import pandas 
pokemon = pandas.read_csv("Pokemon.csv")
```

```python
pokemon.describe()
```

```python
pokemon["Type 1"].value_counts()
```

```python
pokemon.sort_values("Defense",ascending=False)
```

```python
speed_mean = pokemon["Speed"].mean()
def set_speed(inp):
    if inp < speed_mean:
        return "Low speed"
    elif inp == speed_mean:
        return "Fine speed"
    else:
        return "High speed"
```

```python
pokemon["High_low_speed"] = pokemon["Speed"].apply(set_speed)
pokemon
```

```python
hp_mean = pokemon["HP"].mean()
def set_hp(inph):
    if inph < hp_mean:
        return "Low hp"
    elif inph == hp_mean:
        return "Fine Hp"
    else:
        return "High HP"
```

```python
pokemon["High_low_HP"] = pokemon["HP"].apply(set_hp)
pokemon
```
